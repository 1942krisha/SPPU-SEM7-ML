Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def objective(x):
    return (x+3)**2

def derive(x):
    return 2*(x+3)
```

```python
x_list = []
fx_list = []
def gradient(learning_rate, start, iterations):
    x = start
    x_list.append(x)
    fx_list.append(objective(x))
    for i in range(iterations):
        der = derive(x)
        x = x - (der * learning_rate)
        x_list.append(x)
        fx_list.append(objective(x))
        print(f"iter{i+1}, x={round(x,2)}, f(x)={round(objective(x), 2)}")
```

```python
learning_rate = 0.1
start = 2
iterations = 30
gradient(learning_rate, start, iterations)
```

```python
x_val = np.linspace(-15, 15, 100)
plt.plot(x_val, objective(x_val))
plt.plot(2, objective(2), 'ro')
plt.plot(x_list[29], fx_list[29], 'go')
```

```python
x_val = np.linspace(-4, 4, 100)
plt.plot(x_val, objective(x_val))
plt.plot(x_list, fx_list, '.-', color="red")
```
